# 성별 열 나누기

## 성별 분리

### Subtask:
'성별' 열을 기준으로 데이터를 남성('M')과 여성('F')으로 분리합니다.


**Reasoning**:
Filter the DataFrame based on the '성별' column to create separate dataframes for male and female.



```python
df_male = df[df['성별'] == 'M'].copy()
df_female = df[df['성별'] == 'F'].copy()

display(df_male.head())
display(df_female.head())
```

## 열 이름 변경

### Subtask:
분리된 데이터프레임에서 각 연령대별 생활인구수 열의 이름에 성별을 나타내는 접두사를 추가합니다 (예: '0-9생활인구수' -> 'M_0-9생활인구수', 'F_0-9생활인구수').


**Reasoning**:
Rename the age group columns in `df_male` and `df_female` by adding gender prefixes as instructed.



```python
age_columns = [col for col in df_male.columns if '생활인구수' in col]

male_rename_map = {col: 'M_' + col for col in age_columns}
female_rename_map = {col: 'F_' + col for col in age_columns}

df_male = df_male.rename(columns=male_rename_map)
df_female = df_female.rename(columns=female_rename_map)

display(df_male.head())
display(df_female.head())
```

## 데이터 통합

### Subtask:
'연월'과 '행정동코드'를 기준으로 두 데이터프레임을 통합합니다.


**Reasoning**:
Drop the '성별' column from both dataframes and then merge them based on '연월' and '행정동코드'.



```python
df_male = df_male.drop('성별', axis=1)
df_female = df_female.drop('성별', axis=1)

merged_df = pd.merge(df_male, df_female, on=['연월', '행정동코드'], how='outer')
display(merged_df.head())
```

## 결과 확인

### Subtask:
통합된 데이터프레임의 처음 몇 행과 정보 요약을 출력하여 결과가 올바른지 확인합니다.

## 전체 인구 흐름 파악

```python
#  전체 기간 기준 **총생활인구수** 변화 추이 시각화
monthly_population = df.groupby('연월')['총생활인구수'].sum().reset_index()
average_population = monthly_population['총생활인구수'].mean()

plt.figure(figsize=(12, 6))
plt.plot(monthly_population['연월'], monthly_population['총생활인구수'])
plt.axhline(y=average_population, color='r', linestyle='--', label=f'Average: {average_population:.2f}')
plt.title('전체 기간 기준 총생활인구수 변화 추이')
plt.xlabel('연월')
plt.ylabel('총생활인구수')
plt.grid(True)
plt.xticks(monthly_population['연월'], monthly_population['연월'].dt.strftime('%Y-%m'), rotation=45)
plt.legend()
plt.show()
```

### 시군구 기준 순위 확인 : 나이 + 성별

# Task
Calculate and display the ranking of each '시군구명' based on the sum of all columns containing '생활인구수'.

## Data analysis

### Subtask:
For each city/district, calculate the sum of all columns containing '생활인구수'.


**Reasoning**:
Calculate the sum of all population columns for each city/district.



```python
population_columns = [col for col in df.columns if '생활인구수' in col]
sigungu_population_sums = df.groupby('시군구명')[population_columns].sum().reset_index()
display(sigungu_population_sums.head())
```

## Rank cities/districts for each population column

### Subtask:
For each population column, rank the cities/districts based on their summed population in descending order.


**Reasoning**:
For each population column, calculate the rank of each '시군구명' based on the values in that column in descending order and store these rankings in new columns within `sigungu_population_sums`.



```python
for col in population_columns:
  sigungu_population_sums[f'{col}_rank'] = sigungu_population_sums[col].rank(ascending=False)

display(sigungu_population_sums.head())
```

## Combine rankings

### Subtask:
Consolidate the rankings for all population columns into a single, easily readable format.


**Reasoning**:
Select the specified columns from the dataframe and display the result.



```python
ranked_sigungu = sigungu_population_sums[['시군구명'] + [col for col in sigungu_population_sums.columns if col.endswith('_rank')]]
display(ranked_sigungu.head())
```
