```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
import re

# ✅ 크롬 옵션 설정
options = Options()
options.add_argument("--start-maximized")
options.add_experimental_option("detach", True)
options.add_argument("--disable-blink-features=AutomationControlled")

# ✅ 웹드라이버 실행
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# ✅ 행정동 추출 함수
def extract_행정동(address):
    for part in address.split():
        if part.endswith('동'):
            return part
    return None

# ✅ 지번 주소 추출 함수 (강화 버전)
def extract_main_jibun(text):
    lines = text.strip().split('\n')
    for i, line in enumerate(lines):
        if '지번' in line:
            if len(line.strip()) > 2:
                return line.replace('지번', '').strip()
            elif i + 1 < len(lines):
                return lines[i + 1].strip()
    match = re.search(r'지번\s*([^\n]+)', text)
    if match:
        return match.group(1).strip()
    return None

# ✅ 지하철 역명 CSV 로딩
subway = pd.read_csv('sub.csv')  # '역명' 컬럼 필요
subway = subway['역명'].dropna().unique()
total = len(subway)

results = []
start_time = time.time()

# ✅ 크롤링 루프 시작
for i, keyword in enumerate(subway, start=1):
    try:
        # 네이버 지도 검색 접속
        url = f'https://m.map.naver.com/search?query={keyword}역'
        driver.get(url)
        time.sleep(3)  # 초기 로딩 대기

        # 첫 번째 검색 결과 클릭
        driver.find_element(By.XPATH, '//*[@id="ct"]/div/ul/li[1]/div[1]/div[1]/div/div/button').click()
        time.sleep(2)

        # 주소 블록 추출
        full_address = driver.find_element(By.XPATH, '//*[@id="ct"]/div/ul/li[1]/div[1]/div[1]/div[2]/div').text

        # 지번 주소 및 행정동 추출
        지번주소 = extract_main_jibun(full_address)
        행정동 = extract_행정동(지번주소 if 지번주소 else "")

        results.append({
            '역명': keyword,
            '전체주소': 지번주소,
            '행정동': 행정동
        })

        # 진행률 및 예상 시간 출력
        elapsed = time.time() - start_time
        avg_time = elapsed / i
        remaining = avg_time * (total - i)
        print(f"[{i:04}/{total:04} | {i / total * 100:.2f}%] '{keyword}' 완료 → {지번주소 if 지번주소 else 'None'} | 남은시간 ≈ {remaining / 60:.1f}분")

    except Exception as e:
        print(f"[{i:04}/{total:04}] '{keyword}' 오류 발생: {e}")
        results.append({
            '역명': keyword,
            '전체주소': '오류',
            '행정동': None
        })
        continue

    # ✅ 100개마다 중간 저장
    if i % 100 == 0:
        df_temp = pd.DataFrame(results)
        df_temp.to_csv(f'중간저장_{i}.csv', index=False, encoding='utf-8-sig')

# ✅ 최종 저장
df_result = pd.DataFrame(results)
df_result.to_csv('지하철_행정동_결과.csv', index=False, encoding='utf-8-sig')
print("✅ 전체 완료 및 저장됨.")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# Selenium 설정
options = Options()
options.add_argument("--start-maximized")
options.add_experimental_option("detach", True)
options.add_argument("--disable-blink-features=AutomationControlled")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

url = 'https://www.juso.go.kr/openIndexPage.do'
driver.get(url)

# 행정동 추출 함수
def extract_행정동(address):
    for part in address.split():
        if part.endswith('동'):
            return part
    return None

# 지번주소 정규화 함수
def extract_main_jibun(text):
    lines = text.strip().split('\n')
    for i, line in enumerate(lines):
        if '지번' in line:
            if i + 1 < len(lines):
                return lines[i + 1].strip()
    return None

# 상권 데이터 불러오기
juso = pd.read_csv('지하철역.csv')
juso = juso['지하철역']
total = len(juso)

results = []

for i, keyword in enumerate(juso, start=1):
    try:
        # 현재 진행상황 출력
        progress = (i / total) * 100
        print(f"[{i:04}/{total:04} | {progress:.2f}%] '{keyword}' 처리 중...")

        # 검색
        search = driver.find_element(By.XPATH, '//*[@id="mainHeaderSearch"]')
        search.clear()
        search.send_keys(keyword+'역')
        driver.find_element(By.XPATH, '//*[@id="searchButton"]').click()
        time.sleep(2)

        # 주소 추출
        address_list = driver.find_elements(By.CSS_SELECTOR, '#totalResultBox ul li')
        selected_address = None

        for item in address_list:
            if '서울특별시' in item.text:
                try:
                    selected_address = item.find_element(By.CSS_SELECTOR, 'div.roadName').text
                except:
                    selected_address = item.text
                break

        if selected_address:
            # 지번주소 정규화 및 행정동 추출
            jibun = extract_main_jibun(selected_address)
            행정동 = extract_행정동(jibun)
            results.append({
                '지하철역': keyword,
                '전체주소': jibun,
                '행정동': 행정동
            })
            print(f" → 주소 추출 완료: {jibun}")
        else:
            results.append({
                '지하철역': keyword,
                '전체주소': '주소 없음',
                '행정동': None
            })
            print(" → 주소 없음")

    except Exception as e:
        print(f" → 오류 발생: {e}")
        results.append({
            '지하철역': keyword,
            '전체주소': '오류 발생',
            '행정동': None
        })
        continue

# CSV 저장
df_result = pd.DataFrame(results)
df_result.to_csv('지하철역_역주소.csv', index=False, encoding='utf-8-sig')
```
