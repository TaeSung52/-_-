```python
# Create a mapping DataFrame from df_3, excluding the header row
df_mapping_filtered = df_3.iloc[1:].copy()
df_mapping_filtered = df_mapping_filtered.rename(columns={'통계청행정동코드': '통계청행정동코드', '행정동코드': '행정동코드_8자리', '행정동명': '행정동명'})

# Map '출발 행정동 코드' in df_1
df_1['출발 행정동 코드'] = df_1['출발 행정동 코드'].astype(str)
df_1 = df_1.merge(df_mapping_filtered[['통계청행정동코드', '행정동코드_8자리', '행정동명']],
                  left_on='출발 행정동 코드',
                  right_on='통계청행정동코드',
                  how='left',
                  suffixes=('', '_출발'))

df_1 = df_1.rename(columns={'행정동코드_8자리': '출발_행정동_코드_8자리', '행정동명': '출발_행정동_명'})
df_1 = df_1.drop(columns=['통계청행정동코드'])

# Map '도착 행정동 코드' in df_1
df_1['도착 행정동 코드'] = df_1['도착 행정동 코드'].astype(str)
df_1 = df_1.merge(df_mapping_filtered[['통계청행정동코드', '행정동코드_8자리', '행정동명']],
                  left_on='도착 행정동 코드',
                  right_on='통계청행정동코드',
                  how='left',
                  suffixes=('', '_도착'))

df_1 = df_1.rename(columns={'행정동코드_8자리': '도착_행정동_코드_8자리', '행정동명': '도착_행정동_명'})
df_1 = df_1.drop(columns=['통계청행정동코드'])

# Display the first few rows of df_1 with the newly added columns
display(df_1.head())
```

- 시각화를 위한 좌표값도 가져옵니다.

```python
# Create a new DataFrame center_coords from df_2
center_coords = df_2[['코드', 'Y', 'X']].copy()
center_coords = center_coords.rename(columns={'코드': '행정동코드_8자리', 'Y': 'lat', 'X': 'lon'})
center_coords['행정동코드_8자리'] = center_coords['행정동코드_8자리'].astype(str)

# Merge df_1 with center_coords for origin coordinates
df_1 = df_1.merge(center_coords[['행정동코드_8자리', 'lat', 'lon']],
                  left_on='출발_행정동_코드_8자리',
                  right_on='행정동코드_8자리',
                  how='left')
df_1 = df_1.rename(columns={'lat': 'o_lat', 'lon': 'o_lon'})
df_1 = df_1.drop(columns=['행정동코드_8자리'])

# Merge the result with center_coords for destination coordinates
df_1 = df_1.merge(center_coords[['행정동코드_8자리', 'lat', 'lon']],
                  left_on='도착_행정동_코드_8자리',
                  right_on='행정동코드_8자리',
                  how='left')
df_1 = df_1.rename(columns={'lat': 'd_lat', 'lon': 'd_lon'})
df_1 = df_1.drop(columns=['행정동코드_8자리'])

# Display the first few rows of the final merged DataFrame
display(df_1.head())
```

```python
# Identify the columns to sum
mobility_cols = [col for col in df_1.columns if '이동인구(합)' in col]

# Sum the columns and create the '이동인구(합)' column
df_1['이동인구(합)'] = df_1[mobility_cols].sum(axis=1)

# Display the first few rows with the new column
display(df_1.head())
```

```python
def filter_mobility_data(df, year, month):
    """
    Filters the mobility data based on the given year and month.

    Args:
        df: The input DataFrame with a '대상연월' column.
        year: The year for filtering (integer).
        month: The month for filtering (integer).

    Returns:
        A filtered DataFrame containing data for the specified year and month.
    """
    yyyymm = f"{year}{month:02d}"
    filtered_df = df[df['대상연월'] == int(yyyymm)].copy()
    return filtered_df

# Example usage (will not be executed in the final submission, but for testing)
# year_input = 2023
# month_input = 2
# filtered_df = filter_mobility_data(df_1, year_input, month_input)
# print(f"Filtered data for {year_input}-{month_input}:")
# display(filtered_df.head())
# print(f"Shape of filtered data: {filtered_df.shape}")
```

- Top 10만 출력할 경우 맵이 빈약하여 범위를 Top 50으로 확대하여 진행

```python
import pandas as pd

df_1 = pd.read_csv('생활이동_202301_202507_통합_HW_WH.csv')
df_2 = pd.read_csv('서울시_행정동_중심점_2017.csv', encoding='cp949')
df_3 = pd.read_csv('행정동코드_mapping.csv', encoding='cp949')

def analyze_mobility_table(df_mobility, df_district_coords, df_district_mapping, year, month):
    """
    Analyzes mobility data and returns tables of top flows, excluding cases where
    origin and destination administrative district codes are the same.

    Args:
        df_mobility: DataFrame containing mobility data.
        df_district_coords: DataFrame containing administrative district coordinates.
        df_district_mapping: DataFrame containing administrative district mapping.
        year: The year for filtering (integer).
        month: The month for filtering (integer).

    Returns:
        A tuple containing three DataFrames:
        - top_origin: Top 10 origin administrative districts by total mobility (excluding same origin/destination).
        - top_dest: Top 10 destination administrative districts by total mobility (excluding same origin/destination).
        - top_routes: Top 10 routes (origin-destination pairs) by total mobility (excluding same origin/destination).
    """

    # Create copies of the dataframes to avoid modifying the original ones
    df_mobility_copy = df_mobility.copy()
    df_district_coords_copy = df_district_coords.copy()
    df_district_mapping_copy = df_district_mapping.copy()


    # Data Cleaning and Preparation
    # Ensure '대상연월' is in string format for consistent comparison
    df_mobility_copy['대상연월'] = df_mobility_copy['대상연월'].astype(str)
    df_mobility_copy['도착 행정동 코드'] = df_mobility_copy['도착 행정동 코드'].fillna(0).astype(int)

    # Ensure mapping DataFrame has correct column names and types
    df_mapping_filtered = df_district_mapping_copy.iloc[1:].copy()
    df_mapping_filtered = df_mapping_filtered.rename(columns={'통계청행정동코드': '통계청행정동코드', '행정동코드': '행정동코드_8자리', '행정동명': '행정동명'})
    df_mapping_filtered['통계청행정동코드'] = df_mapping_filtered['통계청행정동코드'].astype(str)
    df_mapping_filtered['행정동코드_8자리'] = df_mapping_filtered['행정동코드_8자리'].astype(str)


    # Mapping Administrative District Codes for Origin
    df_mobility_copy['출발 행정동 코드'] = df_mobility_copy['출발 행정동 코드'].astype(str)
    df_mobility_copy = df_mobility_copy.merge(df_mapping_filtered[['통계청행정동코드', '행정동코드_8자리', '행정동명']],
                      left_on='출발 행정동 코드',
                      right_on='통계청행정동코드',
                      how='left')

    df_mobility_copy = df_mobility_copy.rename(columns={'행정동코드_8자리': '출발_행정동_코드_8자리', '행정동명': '출발_행정동_명'})
    df_mobility_copy = df_mobility_copy.drop(columns=['통계청행정동코드'])

    # Mapping Administrative District Codes for Destination
    df_mobility_copy['도착 행정동 코드'] = df_mobility_copy['도착 행정동 코드'].astype(str)
    df_mobility_copy = df_mobility_copy.merge(df_mapping_filtered[['통계청행정동코드', '행정동코드_8자리', '행정동명']],
                      left_on='도착 행정동 코드',
                      right_on='통계청행정동코드',
                      how='left',
                      suffixes=('', '_도착')) # Add suffix for destination merge


    df_mobility_copy = df_mobility_copy.rename(columns={'행정동코드_8자리': '도착_행정동_코드_8자리', '행정동명': '도착_행정동_명'})
    # Dropping the duplicated merge key column after the destination merge
    # Check if the column exists before dropping
    if '통계청행정동코드_도착' in df_mobility_copy.columns:
      df_mobility_copy = df_mobility_copy.drop(columns=['통계청행정동코드_도착'])
    elif '통계청행정동코드' in df_mobility_copy.columns: # Handle case if no suffix was added
        df_mobility_copy = df_mobility_copy.drop(columns=['통계청행정동코드'])


    # Merging with Geographic Data (not strictly necessary for tables, but kept for consistency with previous steps)
    center_coords = df_district_coords_copy[['코드', 'Y', 'X']].copy()
    center_coords = center_coords.rename(columns={'코드': '행정동코드_8자리', 'Y': 'lat', 'X': 'lon'})
    center_coords['행정동코드_8자리'] = center_coords['행정동코드_8자리'].astype(str)

    # Calculating Total Mobility
    mobility_cols = [col for col in df_mobility_copy.columns if '이동인구(합)' in col]
    df_mobility_copy['이동인구(합)'] = df_mobility_copy[mobility_cols].sum(axis=1)

    # Filtering by Year and Month
    yyyymm = f"{year}{month:02d}"
    # Ensure '대상연월' in df_mobility_copy is treated as string for comparison
    filtered_df = df_mobility_copy[df_mobility_copy['대상연월'] == yyyymm].copy()

    # Debugging: Print the shape of the filtered DataFrame and check for the target value
    print(f"Shape of filtered_df for {yyyymm}: {filtered_df.shape}")
    print(f"Does '대상연월' column contain '{yyyymm}'? {yyyymm in df_mobility_copy['대상연월'].unique()}")


    # Exclude rows where origin and destination codes are the same
    filtered_df = filtered_df[filtered_df['출발_행정동_코드_8자리'] != filtered_df['도착_행정동_코드_8자리']].copy()

    # Identifying Top Mobility Flows
    top_origin = filtered_df.groupby(['출발_행정동_코드_8자리', '출발_행정동_명'])['이동인구(합)'].sum().reset_index()
    top_origin = top_origin.sort_values(by='이동인구(합)', ascending=False).head(10)

    top_dest = filtered_df.groupby(['도착_행정동_코드_8자리', '도착_행정동_명'])['이동인구(합)'].sum().reset_index()
    top_dest = top_dest.sort_values(by='이동인구(합)', ascending=False).head(10)

    top_routes = filtered_df.groupby(['출발_행정동_코드_8자리', '출발_행정동_명', '도착_행정동_코드_8자리', '도착_행정동_명'])['이동인구(합)'].sum().reset_index()
    top_routes = top_routes.sort_values(by='이동인구(합)', ascending=False).head(50)

    # Presenting Results (Display tables)
    print(f"--- Mobility Analysis Results for {year}-{month:02d} (Excluding same origin/destination) ---")
    print("\nTop 10 Origin Administrative Districts:")
    display(top_origin)
    print("\nTop 10 Destination Administrative Districts:")
    display(top_dest)
    print("\nTop 50 Routes (Origin-Destination Pairs):")
    display(top_routes)

    return top_origin, top_dest, top_routes
```

```python
# Get year and month input from the user
year_input = int(input("분석할 연도를 입력하세요 (예: 2023): "))
month_input = int(input("분석할 월을 입력하세요 (예: 2월은 2): "))

# Call the analysis function with user input
top_origin_df, top_dest_df, top_routes_df = analyze_mobility_table(df_1, df_2, df_3, year_input, month_input)
```

선 두께: 이동량이 많을수록 두껍게 표시
**주요 관찰 결과:**

*   **주요 이동 거점:** 맵 상에서 선이 집중적으로 모이거나 시작/끝나는 지역은 해당 월에 인구 이동이 활발한 주요 행정동임을 알 수 있습니다. (예: 역삼1동, 여의동, 종로1.2.3.4가동 등)
*   **주요 이동 경로:** 두껍게 표시된 선들은 해당 기간 동안 가장 많은 인구 이동이 발생한 경로를 나타냅니다. (예: 영등포동 <-> 여의동, 신촌동 <-> 연희동, 독산1동 <-> 가산동 등) 이러한 경로는 출퇴근, 상업 활동, 교육 등 다양한 목적의 이동을 반영할 수 있습니다.
*   **지역별 특성:** 특정 지역 내에서의 이동이 두드러지거나 (예: 강남구 내 이동, 마포구 내 이동) 인접한 구 간의 이동이 활발한 패턴을 확인할 수 있습니다.
*   **이동량 분포:** 선의 두께를 통해 각 경로별 인구 이동량의 상대적인 규모를 직관적으로 파악할 수 있습니다.

#### 추가
- html로 맵 구현 후 이동경로를 시각화 할 수 있는 것인 확인되어서 추가해보았습니다.

```python
# -*- coding: utf-8 -*-
# ====== 경로 & 파라미터 ======
TOPO_PATH   = "seoul_municipalities_topo.json"   # 업로드한 TopoJSON
CENT_PATH   = "서울시_행정동_중심점_2017.csv"     # '코드','읍면동명','X','Y'
FLOWS_PATH  = "20_25_타깃_07-10.csv"             # 인구이동 데이터
MAP_PATH    = "행정동코드_mapping.csv"           # 통계청 행정동코드 매핑
DONG_NAME   = "가산동"                           # 대상 동 이름
TOP_N       = 30                                # 상위 N개 흐름만 표시
OUT_HTML    = "flow_map.html"

# ====== 임포트 ======
import json, numpy as np, pandas as pd
import folium
from folium.plugins import AntPath
from branca.colormap import linear
from sklearn.preprocessing import MinMaxScaler
from folium.features import DivIcon

# ====== 공용 유틸 ======
def pick_col(df, cands):
    for c in cands:
        if c in df.columns: return c
    raise KeyError(f"컬럼 없음: {cands}")

def quad_bezier(p0, p1, curv=0.18, n=60):
    x0,y0 = p0; x1,y1 = p1
    mx,my = (x0+x1)/2, (y0+y1)/2
    dx,dy = (x1-x0),(y1-y0)
    nx,ny = -dy,dx
    nrm = (nx**2+ny**2)**0.5 or 1.0
    nx,ny = nx/nrm, ny/nrm
    dist = (dx**2+dy**2)**0.5
    cx,cy = mx + nx*curv*dist, my + ny*curv*dist
    ts = np.linspace(0,1,n)
    pts = []
    for t in ts:
        xt = (1-t)**2*x0 + 2*(1-t)*t*cx + t**2*x1
        yt = (1-t)**2*y0 + 2*(1-t)*t*cy + t**2*y1
        pts.append([xt, yt])
    return pts

def add_label(m, latlon, text, font_size=28, weight=800):
    folium.Marker(
        latlon,
        icon=DivIcon(
            icon_size=(200,36),
            icon_anchor=(0,0),
            html=f'''
            <div style="font-size:{font_size}px;font-weight:{weight};
                        color:#1f2937;text-shadow:0 0 10px rgba(255,255,255,.9),1px 1px 0 #fff;">
                {text}
            </div>'''
        )
    ).add_to(m)

# ====== 중심점/코드 로드 ======
centers = pd.read_csv(CENT_PATH, encoding='cp949')
code_col = pick_col(centers, ["코드","ADM_CD","code"])
name_col = pick_col(centers, ["읍면동명","동명","name"])
lon_col  = pick_col(centers, ["X","lon","경도"])
lat_col  = pick_col(centers, ["Y","lat","위도"])

# 대상 동 좌표
row = centers[centers[name_col]==DONG_NAME]
if len(row)==0:
    raise ValueError(f"'{DONG_NAME}' 중심점을 찾지 못했습니다. CSV를 확인하세요.")
t_lon, t_lat = float(row[lon_col].iloc[0]), float(row[lat_col].iloc[0])
target_code  = str(row[code_col].iloc[0])

# ====== 지도 베이스 ======
m = folium.Map(location=[t_lat, t_lon], zoom_start=14, tiles="CartoDB positron")

# ====== TopoJSON 경계 (대상 동만 하이라이트) ======
with open(TOPO_PATH, encoding="utf-8") as f:
    topo = json.load(f)
obj_key = list(topo["objects"].keys())[0]

def style_fn(feat):
    props = feat.get("properties", {})
    for k in ("EMD_KOR_NM","EMD_NM","adm_nm","읍면동","읍면동명","dong","name"):
        if k in props and str(props[k]) == DONG_NAME:
            return dict(color="#374151", weight=1.5, fillColor="#6b7280", fillOpacity=0.35)
    return dict(color="#9ca3af", weight=0.5, fillOpacity=0.0)

folium.TopoJson(
    topo,
    object_path=f"objects.{obj_key}",
    style_function=style_fn
).add_to(m)

# 중앙 라벨
add_label(m, [t_lat, t_lon], DONG_NAME)

# ====== 흐름 데이터 준비 ======
flows = pd.read_csv(FLOWS_PATH, encoding='cp949')
mapping_df = pd.read_csv(MAP_PATH, encoding='cp949')

# Convert code columns to string for merging
flows['출발 행정동 코드'] = flows['출발 행정동 코드'].astype(str)
flows['도착 행정동 코드'] = flows['도착 행정동 코드'].astype(str)
mapping_df['통계청행정동코드'] = mapping_df['통계청행정동코드'].astype(str)
mapping_df['행정동코드'] = mapping_df['행정동코드'].astype(str)


# 총 이동인구 (남자 20 + 25)
flows['총_이동인구(합)'] = flows['남자_20_이동인구(합)'] + flows['남자_25_이동인구(합)']

# 출발지 매핑
flows = flows.merge(
    mapping_df[['통계청행정동코드','행정동코드','행정동명','시군구명']],
    left_on="출발 행정동 코드", right_on="통계청행정동코드", how="left"
).rename(columns={"행정동명":"출발_행정동명","행정동코드":"출발_행정동코드(행자부)"})

# 도착지 매핑
flows = flows.merge(
    mapping_df[['통계청행정동코드','행정동코드','행정동명','시군구명']],
    left_on="도착 행정동 코드", right_on="통계청행정동코드", how="left"
).rename(columns={"행정동명":"도착_행정동명","행정동코드":"도착_행정동코드(행자부)"})

# ====== 대상 동으로 유입만 필터 ======
flows_to_target = flows[flows["도착_행정동명"]==DONG_NAME].copy()

if flows_to_target.empty:
    print(f"Note: No incoming population movement found for '{DONG_NAME}'.")
else:
    top = flows_to_target.groupby("출발_행정동명", as_index=False)["총_이동인구(합)"].sum() \
                        .sort_values("총_이동인구(합)", ascending=False).head(TOP_N)

    # Get the list of top 2 origin dong names for highlighting
    top_2_origins = top['출발_행정동명'].head(2).tolist()


    # 굵기/색상 스케일
    # Adjusting the color scale to make top values more distinctly blue
    scaler = MinMaxScaler()
    top["w"] = scaler.fit_transform(top[["총_이동인구(합)"]]).ravel()
    # Using a blues scale that starts from a slightly darker blue
    cmap = linear.Blues_09.scale(top["w"].min(), top["w"].max())

    # Code to coordinate mapping
    name2lat = dict(zip(centers[name_col], centers[lat_col]))
    name2lon = dict(zip(centers[name_col], centers[lon_col]))

    # ====== Drawing arcs ======
    for _, r in top.iterrows():
        o_name = r["출발_행정동명"]
        if o_name not in name2lat:
            continue
        o_lat, o_lon = float(name2lat[o_name]), float(name2lon[o_name])
        path = quad_bezier([o_lat, o_lon], [t_lat, t_lon], curv=0.18, n=60)
        w = 3 + r["w"]*8 # Base weight + scaled weight

        # Highlight the top 2 flows with a specific blue color
        if o_name in top_2_origins: # Check if the current origin is in the top 2 list
             col = '#08519c' # A darker blue for highlighting
             w = 5 + r["w"] * 10 # Make the line slightly thicker
        else:
             col = cmap(r["w"]) # Use the scaled color for other flows

        # Shadow
        folium.PolyLine(path, weight=w+6, color="#111827", opacity=0.15).add_to(m)
        # Main line (with animation)
        AntPath(path, weight=w, color=col, opacity=0.85,
                delay=800, dash_array=[10,20]).add_to(m)

    # ====== Save ======
    m.save(OUT_HTML)
    print(f"✅ Saved: {OUT_HTML}")
```
